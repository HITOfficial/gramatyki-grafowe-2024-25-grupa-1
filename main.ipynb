```python
from point import Point, EPoint
from edge import Edge, EEdge
from graph_mk1 import Graph

def b_production(graph, edge):
    graph.remove_edge_prod(edge)
    w = Point(EPoint.v)
    p = edge.attributes[Edge.ATTRIBUTE_P] - 1
    edge_type = EEdge.B if p > 0 else EEdge.A

    graph.add_edge_prod(Edge(edge_type, edge.v, w, {Edge.ATTRIBUTE_P: p}))
    graph.add_edge_prod(Edge(edge_type, w, edge.u, {Edge.ATTRIBUTE_P: p}))

v1 = Point(EPoint.v)
v2 = Point(EPoint.v)

e1 = Edge(EEdge.B, v1, v2, {Edge.ATTRIBUTE_P: 2})

productions = {EEdge.B : b_production}

g = Graph([v1, v2], [e1], productions)
```
